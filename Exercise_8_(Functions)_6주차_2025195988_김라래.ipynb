<h2 id="global">Global variables</h2>

```python
# Example of global variable

artist = "Michael Jackson" #전역변수 = global...
def printer1(artist):
    internal_var = artist #local variable, 즉 printer1이 끝나면 없어지는 것임
    print(artist, "is an artist")

printer1(artist) #artist는 알아도, internal_var 모르기때문에 발생.
```

```python
artist = "Michael Jackson"

def printer(artist):
    global internal_var #일부러 internal_var이 전역이라는 것을 알려줄 수도 있음. 그러나 잘안씀.
    internal_var= "Whitney Houston"
    print(artist,"is an artist")

printer(artist)
printer(internal_var)
```

<h2 id="scope">Scope of a Variable</h2>

```python
# Example of global variable

myFavouriteBand = "AC/DC" #global

def getBandRating(bandname):
    if bandname == myFavouriteBand:
        return 10.0
    else:
        return 0.0

print("AC/DC's rating is:", getBandRating("AC/DC"))
print("Deep Purple's rating is:",getBandRating("Deep Purple"))
print("My favourite band is:", myFavouriteBand)
```

```python
# Example of local variable

def getBandRating(bandname):
    myFavouriteBand2 = "AC/DC" #local variable
    if bandname == myFavouriteBand2:
        return 10.0
    else:
        return 0.0

print("AC/DC's rating is: ", getBandRating("AC/DC"))
print("Deep Purple's rating is: ", getBandRating("Deep Purple"))
print("My favourite band is", myFavouriteBand2)
```

```python
# Example of global variable and local variable with the same name

myFavouriteBand = "AC/DC"

def getBandRating(bandname):
    myFavouriteBand = "Deep Purple"
    if bandname == myFavouriteBand:
        return 10.0
    else:
        return 0.0

print("AC/DC's rating is:",getBandRating("AC/DC"))
print("Deep Purple's rating is: ",getBandRating("Deep Purple"))
print("My favourite band is:",myFavouriteBand)
```
