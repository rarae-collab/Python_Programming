```python
# For loop example

dates = [1982,1980,1973] #List[0] =1982
N = len(dates) #3

for i in range(N): #range(3) = [0, 1, 2] list --> i는 list에 있는 첫번째 값인 0, 그 다음에는 두번째 값인 1, 그 다음에는 세번째 값인 2
    print(dates[i]) # i=0 , i =1 , i =2   --> dates[0] = 1982, dates[1] = 1980, dates[2] =1973
```

```python
# Example of for loop

for i in range(0, 8):
    print("i is equal to", i)
```

```python
# Exmaple of for loop, loop through list

for year in dates:  #이것을 가장 자주 사용한다.
    print(year)
```

```python
# Use for loop to change the elements in list

squares = ['red', 'yellow', 'green', 'purple', 'blue']

for i in range(0, 5):
    print("Before square ", i, 'is',  squares[i])
    squares[i] = 'weight'
    print("After square ", i, 'is',  squares[i])
```

<h3>while loop?

```python
# While Loop Example

dates = [1982, 1980, 1973, 2000]

i = 0
year = 0

while(year != 2000):
    year = dates[i] #1982, 1980, 1973, 2000
    i = i + 1 #i =1, 2, 3, 4
    print(year) #1982, 1980, 1973, 2000

print("It took ", i ,"repetitions to get out of loop.")
```

```python
#Print the elements of the following list: Genres=[ 'rock', 'R&B', 'Soundtrack', 'R&B', 'soul', 'pop']
#Make sure you follow Python conventions.
#The answer is below:

Genres = ['rock', 'R&B', 'Soundtrack', 'R&B', 'soul', 'pop']
for Genre in Genres:
    print(Genre)
```

```python
#Write a for loop that prints out the following list: squares=['red', 'yellow', 'green', 'purple', 'blue']
#The answer is below:

squares=['red', 'yellow', 'green', 'purple', 'blue']
for square in squares:
    print(square)
```

```python
#Write a while loop to display the values of the Rating of an album playlist stored in the list PlayListRatings.
#If the score is less than 6, exit the loop.
#The list PlayListRatings is given by: PlayListRatings = [10, 9.5, 10, 8, 7.5, 5, 10, 10]
#The answer is below:

PlayListRatings = [10, 9.5, 10, 8, 7.5, 5, 10, 10]
i = 1
Rating = PlayListRatings[0]
while(Rating >= 6):
    print(Rating)
    Rating = PlayListRatings[i]
    i = i + 1
```
