## 1. List Comprehension 기초

1부터 5까지의 정수 리스트를 만드는 여러 가지 방법을 확인한다. 먼저 `append()`를 이용하고, 다음으로 `for`문과 `range()`를 이용한다.

```python
# 1부터 5까지의 정수 리스트 만들기 - append 사용
number_list = []
number_list.append(1)
number_list.append(2)
number_list.append(3)
number_list.append(4)
number_list.append(5)

print(number_list)
```

```python
# 1부터 5까지의 정수 리스트 만들기 - for문 사용
number_list = []

for number in range(1, 6):
    number_list.append(number)

print(number_list)
```

```python
# 1부터 5까지의 정수 리스트 만들기 - range와 list 사용
number_list = list(range(1, 6))

print(number_list)
```

## 2. List Comprehension으로 리스트 만들기

리스트 컴프리헨션을 사용하면 `for`문으로 리스트에 값을 추가하는 과정을 한 줄로 작성할 수 있다.

```python
# 기본 List Comprehension
number_list = [number for number in range(1, 6)]
print(number_list)
```

```python
# 함수 적용: 각 숫자에서 1을 빼서 저장
number_list = [number - 1 for number in range(1, 6)]
print(number_list)
```

```python
# 조건 적용: 홀수만 저장
#a_list = [number for number in range(1,6) if number % 2 == 1]
a_list = [number for number in range(1, 6) if number % 2 == 1]
print(a_list)
```

## 3. List Comprehension과 일반 for문 비교

같은 결과를 리스트 컴프리헨션과 일반 `for`문으로 각각 작성해본다.

```python
# List Comprehension 사용
comprehension_list = [number for number in range(1, 6) if number % 2 == 1]
print(comprehension_list)

# 일반 for문 사용
for_list = []
for number in range(1, 6):
    if number % 2 == 1:
        for_list.append(number)

print(for_list)
```

## 4. 문자열 리스트에서 함수와 조건 적용

문자열 리스트에 대해 `lower()` 함수를 적용하고, 단어 길이에 따라 조건을 적용한다. 또한 숫자를 튜플 형태로 저장하는 예제도 실행한다.

```python
sentence = ['I', 'Love', 'Python', 'Soooooo', 'MUCH!!!']

# 함수 적용: 모든 단어를 소문자로 변환
lower_words = [word.lower() for word in sentence]
print(lower_words)

# 조건 적용: 길이가 6보다 큰 단어만 선택
long_words = [word for word in sentence if len(word) > 6]
print(long_words)

# 함수 적용하여 튜플로 저장
power_list = [(x, x ** 2, x ** 3) for x in range(10)]
print(power_list)
```

## 5. 두 개의 for문 사용

중첩된 `for`문을 먼저 실행하고, 같은 결과를 리스트 컴프리헨션으로 작성한다.

```python
# 두 개의 for문 사용
rows = range(1, 4)
cols = range(1, 3)

for row in rows:
    for col in cols:
        print(row, col)
```

```python
# 두 개의 for문을 List Comprehension으로 작성
rows = range(1, 4)
cols = range(1, 3)

cells = [(row, col) for row in rows for col in cols]

for cell in cells:
    print(cell)
```

```python
# 추가 예제: j의 범위가 i에 따라 달라지는 경우
result = [(i, j) for i in range(5) for j in range(i)]
print(result)
```

## 6. Dictionary Comprehension

Dictionary는 `key:value` 형태로 값을 저장한다. 문자열 `letters`에서 각 글자가 몇 번 등장하는지 세어본다.

```python
# Dictionary Comprehension 예제
word = 'letters'
letter_counts = {letter: word.count(letter) for letter in word}

print(letter_counts)
```

```python
# set(word)를 사용하여 중복된 글자를 한 번만 확인
word = 'letters'
letter_counts = {letter: word.count(letter) for letter in set(word)}

print(letter_counts)
```

## 7. Set Comprehension

Set은 중복을 허용하지 않는 자료구조이다. 조건을 만족하는 숫자만 모아 set으로 저장한다.

```python
# Set Comprehension 예제
a_set = {number for number in range(1, 6) if number % 3 == 1}

print(a_set)
```

## 8. Dictionary 뒤집기와 회문 찾기 예제

Dictionary의 key와 value를 바꾸는 예제와, 단어가 앞뒤로 같은지 확인하는 함수를 만들어 set comprehension에 적용한다.

```python
# Dictionary의 key와 value를 바꾸기
d = {'a': 1, 'b': 2, 'c': 3}
reverse_d = {v: k for k, v in d.items()}

print(reverse_d)
```

```python
# 회문 여부를 확인하는 함수
# 예: level, madam처럼 앞에서 읽어도 뒤에서 읽어도 같은 단어

def is_palindrome(word):
    return word == word[::-1]

hamlet = ['level', 'python', 'madam', 'data', 'noon', 'function']
palindrome_words = {word for word in hamlet if is_palindrome(word.lower())}

print(palindrome_words)
```

## 9. 함수 기본 구조

함수는 `def` 키워드로 정의한다. `return`이 있으면 함수 실행 결과를 돌려준다.

```python
# 두 값을 받아서 더한 결과를 반환하는 함수
# 교안의 예제와 같이 여러 자료형에서 동작하는지 확인한다.

def sum(a, b):
    return a + b

print(sum(1, 2))
print(sum(1.3, 3.1))
print(sum('love ', 'python'))
```

## 10. zip()을 이용한 순회

`zip()` 함수는 여러 개의 시퀀스를 병렬로 순회할 때 사용한다. 길이가 다르면 가장 짧은 시퀀스에 맞추어 반복된다.

```python
# zip을 이용한 순회
days = ['Monday', 'Tuesday', 'Wednesday']
fruits = ['banana', 'orange', 'peach']
drinks = ['coffee', 'tea', 'beer']
desserts = ['tiramisu', 'ice cream', 'pie', 'pudding']

for day, fruit, drink, dessert in zip(days, fruits, drinks, desserts):
    print(day, ': drink', drink, '- eat', fruit, '- enjoy', dessert)
```

```python
# zip을 이용하여 영어-프랑스어 사전 만들기
english = 'Monday', 'Tuesday', 'Wednesday'
french = 'Lundi', 'Mardi', 'Mercredi'

print(list(zip(english, french)))
print(dict(zip(english, french)))
```

## 11. 함수 위치 인자

위치 인자는 함수에 값을 넣을 때 순서대로 매개변수에 전달되는 방식이다.

```python
# wine, entree, dessert를 받아서 딕셔너리로 만들어 반환하는 함수

def menu(wine, entree, dessert):
    return {'wine': wine, 'entree': entree, 'dessert': dessert}

result = menu('chardonnay', 'chicken', 'cake')
print(result)
```

## 12. 함수 키워드 인자

키워드 인자는 매개변수 이름을 직접 지정하여 값을 전달하는 방식이다. 순서가 달라도 매개변수 이름에 맞게 전달된다.

```python
# 키워드 인자를 사용한 함수 호출

def menu(wine, entree, dessert):
    return {'wine': wine, 'entree': entree, 'dessert': dessert}

print(menu(entree='beef', dessert='bagel', wine='bordeaux'))
print(menu('frontenac', dessert='flan', entree='fish'))
```

## 13. 기본 매개변수 값 정하기

함수 정의에서 기본값을 정해두면, 함수 호출 시 해당 인자를 생략할 수 있다.

```python
# dessert의 기본값을 정해줌

def menu(wine, entree, dessert='pudding'):
    return {'wine': wine, 'entree': entree, 'dessert': dessert}

print(menu('chardonnay', 'chicken'))
print(menu('dunkelfelder', 'duck', 'doughnut'))
```

## 14. 가능한 함수 호출 예제

기본 매개변수 값을 가진 함수에서 가능한 호출 방식을 실행한다.

```python
def menu(price, wine='chardonnay', entree='chicken', dessert='pudding'):
    return {'price': price, 'wine': wine, 'entree': entree, 'dessert': dessert}

print(menu(100))
print(menu(price=100))
print(menu(price=120, entree='beef'))
print(menu(dessert='bagel', price=110))
print(menu('eighty', 'saint-pierre', 'fish'))
print(menu('hundred', wine='saint-pierre'))
```

## 15. 불가능한 함수 호출 예제

교안의 불가능한 호출은 그대로 실행하면 노트북 전체 실행이 중단된다. 따라서 `try-except`를 사용하여 어떤 오류가 발생하는지 확인한다.

```python
def menu(price, wine='chardonnay', entree='chicken', dessert='pudding'):
    return {'price': price, 'wine': wine, 'entree': entree, 'dessert': dessert}

# price에 인자가 할당되지 않는 경우
try:
    print(menu())
except TypeError as error:
    print('menu() 오류:', error)

# 같은 변수에 두 번 할당하는 경우
try:
    print(menu(100, price=120))
except TypeError as error:
    print('menu(100, price=120) 오류:', error)

# 정의되지 않은 매개변수를 사용하는 경우
try:
    print(menu(main='cream pasta'))
except TypeError as error:
    print("menu(main='cream pasta') 오류:", error)
```

```python
# 키워드 인자 뒤에 위치 인자가 오는 코드는 문법 오류이므로 문자열로 확인한다.
# 실제 코드로 직접 작성하면 노트북 실행이 중단될 수 있다.

bad_code = "menu(price=100, 'saint-pierre')"
print('실행 불가능한 코드 예시:', bad_code)
print('이유: 키워드 인자 뒤에 위치 인자를 쓸 수 없다.')
```

## 16. docstring

함수 시작 부분에 문자열을 작성하면 함수에 대한 설명 문서로 사용할 수 있다.

```python
def echo(anything):
    'echo returns its input argument'
    return anything

print(echo('hello'))
```

```python
def print_if_true(thing, check):
    '''
    Prints the first argument if a second argument is true.
    The operation is:
    1. Check whether the *second* argument is true.
    2. If it is, print the *first* argument.
    '''
    if check:
        print(thing)

print_if_true('Python function practice', True)
print_if_true('This will not be printed', False)
```
