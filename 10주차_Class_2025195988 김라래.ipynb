## 1. 계산기 더하기 기능: 전역변수 사용

처음에는 클래스를 사용하지 않고, 전역변수 `result`에 값을 계속 더하는 방식으로 계산기 기능을 구현한다.

```python
# p.11 계산기 더하기 기능
# 전역변수 result에 숫자를 계속 더하는 함수

result = 0

def adder(num):
    global result
    result += num
    return result

print(adder(3))
print(adder(4))

# 이어서 한 번 더 실행해보면 이전 결과가 계속 저장되어 있음을 확인할 수 있다.
print(adder(4))
print(adder(5))
```

## 2. 계산기가 두 개 필요한 경우

계산기가 두 개라면 결과를 저장하는 변수와 함수도 각각 따로 만들어야 한다. 이 방식은 계산기가 많아질수록 코드가 반복된다.

```python
# p.12 계산기가 두 개 필요한 경우

result1 = 0
result2 = 0

def adder1(num):
    global result1
    result1 += num
    return result1

def adder2(num):
    global result2
    result2 += num
    return result2

print(adder1(3))
print(adder1(4))
print(adder2(3))
print(adder2(7))
```

## 3. 클래스를 사용한 계산기 여러 개 만들기

`Calculator` 클래스를 만들면 같은 구조를 가진 계산기 객체를 여러 개 만들 수 있다. `cal1`, `cal2`는 서로 독립적으로 결과를 저장한다.

```python
# p.13 클래스 사용, 계산기 객체 여러 개 생성

class Calculator:
    def __init__(self):
        self.result = 0

    def adder(self, num):
        self.result += num
        return self.result

cal1 = Calculator()
cal2 = Calculator()

print(cal1.adder(3))
print(cal1.adder(4))
print(cal2.adder(3))
print(cal2.adder(7))
```

## 4. 객체와 인스턴스 확인

클래스로 만든 실제 객체를 인스턴스라고도 한다. `cal1`은 객체이며, `Calculator` 클래스의 인스턴스이다.

```python
# p.14 객체와 인스턴스 확인

print(type(cal1))
print(isinstance(cal1, Calculator))
print(isinstance(cal2, Calculator))
```

## 7. 클래스 내부의 함수: 메서드

클래스 안에서 정의한 함수는 메서드라고 부른다. 메서드의 첫 번째 매개변수는 보통 `self`를 사용한다.

```python
# p.17 클래스 내부의 함수(메서드)

class Service:
    secret = "지구는 4006년에 멸망한다"

    def sum(self, a, b):
        result = a + b
        print("%s + %s = %s이다." % (a, b, result))

an = Service()
an.sum(1, 1)
```

## 8. self의 의미

메서드를 호출할 때 `an.sum(1, 1)`처럼 사용하면, `self`에는 호출한 객체 `an`이 자동으로 전달된다. 그래서 실제 호출 시에는 `self`를 제외한 값만 입력한다.

```python
# p.18 self 이해를 위한 추가 확인

print("an은 Service의 인스턴스인가?", isinstance(an, Service))
print("메서드 호출 결과:")
an.sum(2, 3)
```

## 9. 객체 변수

객체 변수는 객체마다 따로 저장되는 변수이다. 아래 예제에서는 `setname()` 메서드를 통해 객체 변수 `name`을 만든다.

```python
# p.19 객체 변수

class Service:
    secret = "지구는 4006년에 멸망한다"

    def setname(self, name):
        self.name = name

    def sum(self, a, b):
        result = a + b
        print("%s님, %s + %s = %s입니다." % (self.name, a, b, result))

an = Service()
an.setname("박달도사")
an.sum(1, 1)
```

```python
# p.20 클래스 변수와 객체 변수

kim = Service()
park = Service()

kim.name = "김정보"
park.name = "박융합"

print(kim.name)
print(park.name)

# kim 객체에만 secret 객체 변수를 따로 만든 경우
kim.secret = "비밀은 없다"

print(park.secret)   # park는 클래스 변수 값을 사용
print(kim.secret)    # kim은 객체 변수 값을 사용
print(Service.secret) # 클래스 변수 값 확인
```

## 11. `__init__`이 필요한 이유

`setname()`을 호출하지 않으면 `name` 객체 변수가 만들어지지 않아 오류가 발생한다. 실제 과제 파일 전체 실행이 멈추지 않도록 `try-except`로 오류 내용을 확인한다.

```python
# p.21 setname을 호출하지 않는 경우의 문제 확인

lee = Service()

try:
    lee.sum(1, 1)
except AttributeError as e:
    print("오류 발생:", e)
```

## 16. 사칙연산 클래스: 더하기 기능

저장된 두 수를 더하는 `sum()` 메서드를 추가한다.

```python
# p.26 FourCal 더하기 기능

class FourCal:
    def setdata(self, first, second):
        self.first = first
        self.second = second

    def sum(self):
        result = self.first + self.second
        return result

a = FourCal()
a.setdata(4, 2)
print(a.sum())
```

## 17. 사칙연산 클래스 완성

`sum`, `mul`, `sub`, `div` 메서드를 추가하여 덧셈, 곱셈, 뺄셈, 나눗셈을 실행한다.

```python
# p.27 FourCal 사칙연산 클래스

class FourCal:
    def setdata(self, first, second):
        self.first = first
        self.second = second

    def sum(self):
        result = self.first + self.second
        return result

    def mul(self):
        result = self.first * self.second
        return result

    def sub(self):
        result = self.first - self.second
        return result

    def div(self):
        result = self.first / self.second
        return result

a = FourCal()
b = FourCal()

a.setdata(4, 2)
b.setdata(3, 7)

print(a.sum())
print(a.mul())
print(a.sub())
print(a.div())
print(b.sub())
```

## 18. 간단한 주사위 게임 설명

두 명의 플레이어가 각각 주사위 3개를 굴리고, 주사위 합계가 큰 사람이 이기는 게임이다.

## 19. 간단한 주사위 게임: 순차적 프로그래밍

클래스를 사용하지 않고 리스트와 반복문만 사용해 주사위 게임을 작성한다.

```python
# p.30 간단한 주사위 게임 - 순차적 프로그래밍

import random

player1_dice = []
player2_dice = []

for i in range(3):
    player1_dice.append(random.randint(1, 6))
    player2_dice.append(random.randint(1, 6))

print("Player 1 rolled" + str(player1_dice))
print("Player 2 rolled" + str(player2_dice))

if sum(player1_dice) == sum(player2_dice):
    print("Draw")
elif sum(player1_dice) > sum(player2_dice):
    print("Player 1 wins")
else:
    print("Player 2 wins")
```

## 20. 간단한 주사위 게임: OOP 방식

`Player` 클래스를 만들고, 각 플레이어가 자신의 주사위 값을 객체 변수 `dice`에 저장하도록 작성한다.

```python
# p.31 간단한 주사위 게임 - OOP

from random import randint

class Player:
    def __init__(self):
        self.dice = []

    def roll(self):
        self.dice = []  # 현재 주사위 값을 비우고 다시 굴림
        for i in range(3):
            self.dice.append(randint(1, 6))

    def get_dice(self):
        return self.dice

player1 = Player()
player2 = Player()

player1.roll()
player2.roll()

print("Player 1 rolled" + str(player1.get_dice()))
print("Player 2 rolled" + str(player2.get_dice()))

if sum(player1.get_dice()) == sum(player2.get_dice()):
    print("Draw!")
elif sum(player1.get_dice()) > sum(player2.get_dice()):
    print("Player 1 wins!")
else:
    print("Player 2 wins!")
```
