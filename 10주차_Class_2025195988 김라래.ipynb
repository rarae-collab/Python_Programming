## 1. 계산기 더하기 기능: 전역변수 사용

처음에는 클래스를 사용하지 않고, 전역변수 `result`에 값을 계속 더하는 방식으로 계산기 기능을 구현한다.

```python
result = 0
```

```python
def adder(num):
    global result
    result += num
    return result
```

```python
print(adder(3))
```

```python
print(adder(4))
```

```python
print(adder(5))
```

## 3. 클래스를 사용한 계산기 여러 개 만들기

`Calculator` 클래스를 만들면 같은 구조를 가진 계산기 객체를 여러 개 만들 수 있다. `cal1`, `cal2`는 서로 독립적으로 결과를 저장한다.

```python
class Calculator:
    def __init__(self):
        self.result = 0

    def adder(self, num):
        self.result += num
        return self.result
```

```python
cal1 = Calculator()
```

```python
cal1.adder(3)
```

```python
cal1.adder(4)
```

```python
cal2 = Calculator()
```

```python
cal2.adder(3)
```

```python
cal2.adder(7)
```

## 10. `__init__`이 필요한 이유

`setname()`을 호출하지 않으면 `name` 객체 변수가 만들어지지 않아 오류가 발생한다. 실제 과제 파일 전체 실행이 멈추지 않도록 `try-except`로 오류 내용을 확인한다.

```python
# p.21 setname을 호출하지 않는 경우의 문제 확인

lee = Service()

try:
    lee.sum(1, 1)
except AttributeError as e:
    print("오류 발생:", e)
```

## 14. 사칙연산 클래스: 더하기 기능

저장된 두 수를 더하는 `sum()` 메서드를 추가한다.

```python
# p.26 FourCal 더하기 기능

class FourCal:
    def setdata(self, first, second):
        self.first = first
        self.second = second

    def sum(self):
        result = self.first + self.second
        return result

a = FourCal()
a.setdata(4, 2)
print(a.sum())
```

## 15. 사칙연산 클래스 완성

`sum`, `mul`, `sub`, `div` 메서드를 추가하여 덧셈, 곱셈, 뺄셈, 나눗셈을 실행한다.

```python
# p.27 FourCal 사칙연산 클래스

class FourCal:
    def setdata(self, first, second):
        self.first = first
        self.second = second

    def sum(self):
        result = self.first + self.second
        return result

    def mul(self):
        result = self.first * self.second
        return result

    def sub(self):
        result = self.first - self.second
        return result

    def div(self):
        result = self.first / self.second
        return result

a = FourCal()
b = FourCal()

a.setdata(4, 2)
b.setdata(3, 7)

print(a.sum())
print(a.mul())
print(a.sub())
print(a.div())
print(b.sub())
```

## 16. 간단한 주사위 게임 설명

두 명의 플레이어가 각각 주사위 3개를 굴리고, 주사위 합계가 큰 사람이 이기는 게임이다.

## 17. 간단한 주사위 게임: 순차적 프로그래밍

클래스를 사용하지 않고 리스트와 반복문만 사용해 주사위 게임을 작성한다.

```python
# p.30 간단한 주사위 게임 - 순차적 프로그래밍

import random

player1_dice = []
player2_dice = []

for i in range(3):
    player1_dice.append(random.randint(1, 6))
    player2_dice.append(random.randint(1, 6))

print("Player 1 rolled" + str(player1_dice))
print("Player 2 rolled" + str(player2_dice))

if sum(player1_dice) == sum(player2_dice):
    print("Draw")
elif sum(player1_dice) > sum(player2_dice):
    print("Player 1 wins")
else:
    print("Player 2 wins")
```

## 18. 간단한 주사위 게임: OOP 방식

`Player` 클래스를 만들고, 각 플레이어가 자신의 주사위 값을 객체 변수 `dice`에 저장하도록 작성한다.

```python
# p.31 간단한 주사위 게임 - OOP

from random import randint

class Player:
    def __init__(self):
        self.dice = []

    def roll(self):
        self.dice = []  # 현재 주사위 값을 비우고 다시 굴림
        for i in range(3):
            self.dice.append(randint(1, 6))

    def get_dice(self):
        return self.dice

player1 = Player()
player2 = Player()

player1.roll()
player2.roll()

print("Player 1 rolled" + str(player1.get_dice()))
print("Player 2 rolled" + str(player2.get_dice()))

if sum(player1.get_dice()) == sum(player2.get_dice()):
    print("Draw!")
elif sum(player1.get_dice()) > sum(player2.get_dice()):
    print("Player 1 wins!")
else:
    print("Player 2 wins!")
```
